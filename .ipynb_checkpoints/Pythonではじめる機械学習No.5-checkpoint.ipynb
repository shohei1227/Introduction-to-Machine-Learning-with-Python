# 教師なし学習と前処理

## 教師なし学習

データセットの変換と、クラスタリング  
  
データセットの教師なし変換の利用法として最も一般的なのは、次元削減である。  
一方、クラスタリングアルゴリズムはデータを似たような要素から構成されるグループに分けるアルゴリズムである。  
  
教師なし学習は出力が問うあるべきかわからないため、有用性の評価に難しさがある。そのため、データをよりよく理解するために、探索的に用いたり、教師あり学習の前処理ステップとしての利用として使われたりする。

## 前処理とスケール変換

ニューラルネットワークやSVMなどのアルゴリズムはデータのスケール変換に非常に敏感である。したがって、これらのアルゴリズムに適したデータ表現に変換することが広く行われている。

```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd 
import mglearn
from IPython.display import display
import warnings
%matplotlib inline

warnings.simplefilter('ignore')
```

```python
mglearn.plots.plot_scaling()#様々なスケール変換
```

```python
from sklearn.datasets import load_breast_cancer#cancerデータセット
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=1)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)#transformメソッドで実際にスケール変換（学習した変換を適用）

print("変換後の配列の形:{}".format(X_train_scaled.shape))
print("変換前の各特徴量の最小値:{}".format(X_train.min(axis=0)))
print("変換前の各特徴量の最大値:{}".format(X_train.max(axis=0)))
print("変換後の各特徴量の最小値:{}".format(X_train_scaled.min(axis=0)))
print("変換後の各特徴量の最大値:{}".format(X_train_scaled.max(axis=0)))
```

```python
X_test_scaled = scaler.transform(X_test)#テストデータを変換

print("変換後の各特徴量の最小値:{}".format(X_test_scaled.min(axis=0)))
print("変換後の各特徴量の最大値:{}".format(X_test_scaled.max(axis=0)))
#常に訓練データとテストデータに全く同じ変換を施すため、テストセットの場合には、スケール変換後の最小値と最大値が0と1になっていない。
```

```python
from sklearn.datasets import make_blobs
X, _ = make_blobs(n_samples=50, centers=5, random_state=4, cluster_std=2)
X_train, X_test = train_test_split(X, random_state=5, test_size=.1)

fig, axes = plt.subplots(1, 3, figsize=(13, 4))
axes[0].scatter(X_train[:,0], X_train[:,1], c=mglearn.cm2(0), label="Training set", s=60)
axes[0].scatter(X_test[:,0], X_test[:,1],marker="^", c=mglearn.cm2(1), label="Test set", s=60)
axes[0].legend(loc="upper left")
axes[0].set_title("Original Data")

scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

axes[1].scatter(X_train_scaled[:,0], X_train_scaled[:,1], c=mglearn.cm2(0), label="Training set", s=60)
axes[1].scatter(X_test_scaled[:,0], X_test_scaled[:,1], marker="^", c=mglearn.cm2(1), label="Test set", s=60)
axes[1].set_title("Scalded Data")



##実際にやってはいけない（テストセットと訓練セットは同じスケール変換をしなくちゃいけないから）
test_scaler = MinMaxScaler()
test_scaler.fit(X_test)
X_test_scaled_badly = test_scaler.transform(X_test)

axes[2].scatter(X_train_scaled[:,0], X_train_scaled[:,1], c=mglearn.cm2(0), label="Training set", s=60)
axes[2].scatter(X_test_scaled_badly[:,0], X_test_scaled_badly[:,1], marker="^", c=mglearn.cm2(1), label="Test set", s=60)
axes[2].set_title("Improperly Scalded Data")



for ax in axes:
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
```

最初のパネルは変換されていない2次元のデータセットである。  
真ん中は訓練セットに対してスケール変換を行い（そのため、グラフのメモリが変わっている。）、テストセットに対しても全く同じ変換を行ったものである。  
1番左は訓練セットとテストセットを別々に変換した場合を示している。この場合特徴量の最小値最大値は訓練セットのテストセットも0、1になっているが、データセットが変わってしまっている。データを勝手に動かしてしまっているため、まずい。

## 次元削減、特徴量抽出、多様体学習

### 主成分分析(principal component analysis : PCA)

主成分分析とは、データセットの特徴量を相互に統計的に関連しないように回転する手法。多くの場合、回転した後の特徴量から、データを説明するのに重要な一部の特徴量だけを抜き出す。

```python
mglearn.plots.plot_pca_illustration()
```

まずは、最も分散が大きい方向を見つけ、「第１成分」というラベルをつける。
次に、第１成分と直交する方向の中から、最も情報を持っている方向を探し、「第２成分」とする。

左上：元のデータセット  
右上：変換後のデータ  
左下：変換して第２成分を落としたデータ  
右下：第１成分だけを用いて逆回転したデータ  


cancerデータセットのPCAによる可視化

```python
#元のcancerデータセットに対するクラスごとのヒストグラム
fig, axes = plt.subplots(15, 2, figsize = (10, 20))
malignant = cancer.data[cancer.target == 0]
benign = cancer.data[cancer.target == 1]

ax = axes.ravel()

for i in range(30):
    _, bins = np.histogram(cancer.data[:, i], bins=50)
    ax[i].hist(malignant[:, i], bins=bins, color=mglearn.cm3(0),alpha=.5)
    ax[i].hist(benign[:, i], bins=bins, color=mglearn.cm3(2),alpha=.5)
    ax[i].set_title(cancer.feature_names[i])
    ax[i].set_yticks(())
ax[0].set_xlabel("Features magnitude")
ax[0].set_ylabel("Frequency")
ax[0].legend(["malignant", "benign"], loc = "best")
fig.tight_layout()
```

```python
#PCAを適用する前にデータをStandardScalerで特徴量の分散が１になるようにする。
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

scaler = StandardScaler()
scaler.fit(cancer.data)
X_scaled = scaler.transform(cancer.data)
```

```python
#PCAオブジェクトを生成し、fitメソッドを呼び出し、主成分を見つけ、transform メソッドを読んで回転と次元削減を行う
#PCAオブジェクトを作る際に、維持する主成分の数を指定する必要がある。
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca.fit(X_scaled)

X_pca = pca.transform(X_scaled)
print("Original shape:{}".format(str(X_scaled.shape)))#もとの形
print("Reduced shape:{}".format(str(X_pca.shape)))#削減後の形
```

```python
#最初の２つの主成分に対して、２次元でプロットできる。
plt.figure(figsize=(8, 8))
mglearn.discrete_scatter(X_pca[:, 0],X_pca[:, 1], cancer.target)
plt.legend(cancer.target_names, loc="best")
plt.gca().set_aspect("equal")
plt.xlabel("First principal component")
plt.ylabel("second principal component")
```

mean rudius, mean textureの2つの成分

```python
print("PCA component shape: {}".format(pca.components_.shape))
```

```python
print("PCA components:\n{}".format(pca.components_))
```

```python
plt.matshow(pca.components_, cmap="viridis")
plt.yticks([0, 1], ["First component", "Second component"])
plt.colorbar()
plt.xticks(range(len(cancer.feature_names)), cancer.feature_names, rotation=60, ha="left")
plt.xlabel("Feature")
plt.ylabel("Principal components")
```

```python
from sklearn.datasets import fetch_lfw_people
people = fetch_lfw_people(min_faces_per_person=20, resize=0.7)
image_shape = people.images[0].shape

fox, axes = plt.subplots(2, 5, figsize=(15, 8), subplot_kw={"xticks":(), "yticks":()})

for target, image, ax in zip(people.target, people.images, axes.ravel()):
    ax.imshow(image)
    ax.set_title(people.target_names[target])
```

```python
print("People.images.shape:{}".format(people.images.shape))# 全部で３０２３枚、サイズは８７×６５
print("Number of classes:{}".format(len(people.target_names)))#クラスの数、つまり６２人分
```

```python
counts = np.bincount(people.target)
for i, (counts, name) in enumerate(zip(counts, people.target_names)):
    print("{0:30}{1:3}　".format(name, counts), end = " ")
    if(i + 1) % 3 == 0:
        print()
```

George W Bush とColin Powell　の枚数が多く偏っている。

```python
#偏りをなくすため、各人の画像の数を50に制限する。
mask = np.zeros(people.target.shape, dtype=np.bool)
#print(people.target[:10])
for target in np.unique(people.target):
    #print(target)
    mask[np.where(people.target == target)[0][:50]] = 1

    X_people = people.data[mask]
    y_people = people.target[mask]

#print(87 * 65)
#print(X_people.shape)
#print(X_people[1].shape)#すべての画素をRGBで表している
#0～255までで表現されている、グレースケールの値を0と1の間に変換
X_people = X_people / 255.
#print(X_people)
```

```python
#1-最近傍法クラス分類器を用いてクラス分類してみる
from sklearn.neighbors import KNeighborsClassifier

X_train, X_test, y_train, y_test = train_test_split(X_people, y_people, stratify=y_people, random_state=42)

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
print("Test set score of 1_nn: {:.2f}".format(knn.score(X_test, y_test)))

#62クラスの分類と考えると、（ランダムで分類すれば1.5%であるため、）それほど低くはない。
#精度はよくはない
```

```python
mglearn.plots.plot_pca_whitening()
```

左：元のデータ  
右：whitenオプション付きでPCA変換  
（whitehオプションは主成分が同じスケールになるようにスケール変換する。PCA変換後にStandardScalerをかけるのと一緒）

```python
pca = PCA(n_components=100, whiten=True, random_state=0).fit(X_train)
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)

print("X_train_pca.shape:{}".format(X_train_pca.shape))

#新たにPCAオブジェクトを訓練したデータは100の特徴量を持つ
```

```python
#この新なデータで1-最近傍法クラス分類器を用いると…
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train_pca, y_train)
print("Test set accuracy:{:.2f}".format(knn.score(X_test_pca, y_test)))
```

例えば1つ目の主成分は顔の背景のコントラスト、2つ目は光の当たり方による左右の明るさの差...など

PCA変換は回転させてから、分散が小さい成分を落とすものであると紹介してきたが、もう1つ有用な解釈があり、  
PCAはテストデータポイントを主成分の重み付き和として表現する。

このように最初の2つの成分だけでは、データ全体がひとまとまりになってしまい、クラスを分類できそうにない。  
なぜなら、10の主成分を用いてもPCAでは顔の非常に粗い特徴しか捉えられないからである。

### 非負値行列因子分解(Non-negative matrix factorization : NMF)

有用な特徴量を抽出することを目的とする教師なし学習手法。アルゴリズムの動作はPCAと似ており、次元削減に用いることができる。  
成分とその係数が常に0以上であることがも求められる。そのため、この方法は個々の特徴量が非負のデータに対してしか適用できない。  
非負の成分を非負係数で重み付き和をとると非負にしかならないから。

```python
mglearn.plots.plot_nmf_illustration()
```

左：NMFで見つかった2つの特徴量  
右：NMFで見つかった1つの特徴量

```python
mglearn.plots.plot_nmf_faces(X_train, X_test, image_shape) 
```

PCAは再構成に最適な方向を見つけるのに対し、NMFはデータ中から興味深いパターンを見つけるのに用いられるため、上の逆変換したデータ質はPCAを用いたときよりも少し悪くなっている。

```python
#最初の15成分
from sklearn.decomposition import NMF

nmf = NMF(n_components=15, random_state=0)
nmf.fit(X_train)
X_train_nmf = nmf.transform(X_train)
X_test_nmf = nmf.transform(X_test)

fix, axes = plt.subplots(3, 5, figsize=(15, 12), subplot_kw={"xticks": (), "yticks": ()})
for i , (component, ax) in enumerate(zip(nmf.components_, axes.ravel())):
    ax.imshow(component.reshape(image_shape))
    ax.set_title("{}.component".format(i))
```

成分がすべて正になるため、PCAで得られた成分よりもはるかに顔のプロトタイプをとらえている。

```python
compn = 12#12番目の成分でソート、最初の10画像を表示
inds = np.argsort(X_train_nmf[:, compn])[::-1]
fig, axes = plt.subplots(2,5, figsize=(15,8), subplot_kw={"xticks" : (), "yticks" :()})
for i ,(ind, ax) in enumerate(zip(inds, axes.ravel())):
    ax.imshow(X_train[ind].reshape(image_shape))
```

```python
compn = 14#14番目の成分でソート、最初の10画像を表示
inds = np.argsort(X_train_nmf[:, compn])[::-1]
fig, axes = plt.subplots(2,5, figsize=(15,8), subplot_kw={"xticks" : (), "yticks" :()})
for i ,(ind, ax) in enumerate(zip(inds, axes.ravel())):
    ax.imshow(X_train[ind].reshape(image_shape))
```

```python
S = mglearn.datasets.make_signals()#3つの信号源からの信号が組み合わされた信号
plt.figure(figsize=(12, 2))
plt.plot(S, "-")
plt.xlabel("Time")
plt.ylabel("Signal")
plt.title("Original Signal source")#元の信号源
```

これを分解して元の成分を取り出したい。

```python
#データを混ぜて100次元の状態を作る
A = np.random.RandomState(0).uniform(size = (100, 3))
X = np.dot(S, A.T)
print("測定データの形状: {}".format(X.shape))
```

```python
#NMFで復元
nmf = NMF(n_components=3, random_state=42)
S_ = nmf.fit_transform(X)
print("再構成された元信号データの形状：{}".format(S_.shape))
```

```python
#比較のため、PCAでも復元
pca = PCA(n_components=3)
H = pca.fit_transform(X)
print("PCAの形状：{}".format(H.shape))
```

```python
models = [X, S, S_, H]
names = ["Observations (first three meansurements)", "True sorces", "NMF recovered signals", "PCA recovered signals"]
fig, axes = plt.subplots(4, figsize = (20,10), gridspec_kw={"hspace": .5},subplot_kw={"xticks": (), "yticks": ()})

for model, name, ax in zip(models, names, axes):
    ax.set_title(name)
    ax.plot(model[:, :3], "-")
```

1番目：観測データ（１００個の計測結果のうちの3つ）  
2番目：実際の信号  
3番目：NMFによる信号の復元  
4番目：PCAによる信号の復元

NMFはもとの信号源をうまく復元できている。

### t-SNEを用いた多様体学習

データを変換して散布図で可視化したいとき、まずPCAで試してみるのは悪くないが、この手法の回転していくつかの方向を落とすという性質上、その有用性は限られる。これに対して、多様体学習アルゴリズム(manifold learning algorithms)と呼ばれる、一連のアルゴリズムがある。  
これは、はるかに複雑なマッピングを行い、より良い可視性を実現できる。特に有用なのが、t-SNEアルゴリズムである。  
  
多様体学習アルゴリズムはテストセットに適用することはできない。探索的なデータ解析に有用だが、最終的な目標が教師あり学習の場合には、ほとんど用いられない。
  
t-SNEはデータポイントの距離を可能な限り、維持する2次元表現を見つけようとする。  
まず最初にランダムな、2次元表現を作り、そこから、元の特徴空間で近いデータポイントを近くに、遠いデータポイントを遠くに配置しようとする。つまり、どの点が近傍がを示す情報を維持しようとする。
  
  ここでは、scikit-learnの手書きデータセット（digitsデータセット）に適用してみる。

```python
#手書きデータセットの画像例
from sklearn.datasets import load_digits
digits = load_digits()

fig, axes = plt.subplots(1, 10, figsize=(10, 5), subplot_kw={"xticks": (), "yticks":()})
for ax, img in zip(axes.ravel(), digits.images):
    ax.imshow(img)
```

```python
#pcaモデルを構築
pca = PCA(n_components=2)#最初の2成分で変形させる
pca.fit(digits.data)

digits_pca = pca.transform(digits.data)
colors = ["#476A2A", "#7851B8", "#BD3430", "#4A2D4E", "#875525", "#A83683", "#4E655E", "#853541", "#3A3120","#535D8E" ]

plt.figure(figsize=(10, 10))
plt.xlim(digits_pca[:, 0].min(), digits_pca[:, 0].max())
plt.ylim(digits_pca[:, 1].min(), digits_pca[:, 1].max())

for i in range(len(digits.data)):
    plt.text(digits_pca[i, 0], digits_pca[i, 1], str(digits.target[i]), color = colors[digits.target[i]], 
             fontdict={"weight":"bold", "size": 9})
    plt.xlabel("First principal component")
    plt.ylabel("Second principal component")
```

最初の2成分でうまく分離できている数字もあるが、重なり合っている箇所も非常に多い。

t-SNEを適用する。  
t-SNEは新しいデータの変換をサポートしていないため、TSNEクラスには、transformメソッドがない。そのため、fit_transformメソッドを用いる。

```python
from sklearn.manifold import TSNE
tsne = TSNE(random_state=42)
digits_tsne = tsne.fit_transform(digits.data)
```

```python
plt.figure(figsize=(10, 10))
plt.xlim(digits_tsne[:, 0].min(), digits_tsne[:, 0].max() + 1)
plt.ylim(digits_tsne[:, 1].min(), digits_tsne[:, 1].max() + 1)

for i in range(len(digits.data)):
    plt.text(digits_tsne[i, 0], digits_tsne[i, 1], str(digits.target[i]), color = colors[digits.target[i]], 
             fontdict={"weight":"bold", "size": 9})#点ではなく、数字を
    plt.xlabel("t-SNE feature 0")
    plt.ylabel("t-SNE feature 1")
```

かなり良く分離できている。  
クラスラベルの知識を全く使っていない。これは完全に教師なし学習である。
